Step 7) Genre names are stored in array.We need to standarize them inorder to make it easier to read and analyze.(Data Cleaning)

```python
import pandas as pd

def clean_genre_string(genre_str):
    try:
        genre_str = genre_str.strip('[]')
        genres = [g.strip().strip("'").strip('"') for g in genre_str.split(',') if g.strip()]
        return ', '.join(genres)
    except:
        return genre_str if isinstance(genre_str, str) else ''

df['genre_names'] = df['genre_names'].apply(clean_genre_string)
df





```

Step 8)We can remove genre Id from the dataframe,as have extracted the genre names from it.(Data Cleaning)

```python
df = df.drop(columns=['genre_ids'])
df
```

Step 9)We can remove duplicate rows from the dataset(First we count the number of duplicate row and then we delete it)(Data Cleaning)

```python
duplicates_mask = df.duplicated()
num_duplicates = duplicates_mask.sum()
print(f"Number of duplicate rows: {num_duplicates}")
duplicates = df[duplicates_mask]
print("Duplicate Rows:")
print(duplicates)
df
```

Step 10) We need to standardize the genre name column as it would be helpful for analysis and for visualisation.(Data Cleaning)

```python
df['genre_names'] = df['genre_names'].str.lower().str.strip()
```

Step 11) We need to remove Outlier from popularity column.So the best way to remove is IQR method.First identify the outliers and replaice it with lower bounds and upper bounds(Data Cleaning)

```python
initial_rows = df.shape[0]
Q1 = df['popularity'].quantile(0.25)
Q3 = df['popularity'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['popularity'] >= lower_bound) & (df['popularity'] <= upper_bound)]

final_rows = df.shape[0]
df = df.reset_index(drop=True)


print(initial_rows,final_rows)
```
